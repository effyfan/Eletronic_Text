```python
import spacy
import annoy
```

```python
nlp = spacy.load('en_core_web_md')
```

```python
type(nlp.vocab)
```

```python
qualified = [item for item in nlp.vocab if item.has_vector and item.is_alpha]
```

```python
type(qualified)
```

```python
len(qualified)
```

```python
lexmap = []
t = annoy.AnnoyIndex(300)
for i, item in enumerate(islice(sorted(qualified, key=lambda x: x.prob, reverse=True), 100000)):
    t.add_item(i, item.vector)
    lexmap.append(item)
t.build(25)
```

```python
t
```

```python
len(lexmap)
```

```python
def similarsemantic(t, nlp, word, n):
    seen = set()
    count = 0
    for i in t.get_nns_by_vector(nlp.vocab[word].vector, 100):
        this_word = lexmap[i].orth_.lower()
        if this_word not in seen and word != this_word:
            seen.add(this_word)
            count += 1
            yield this_word
            if count >= n:
                break
list(similarsemantic(t, nlp, "hello", 20))
```

```python
chance = 50
nuttiness = 10
doc = nlp(text)
output = []
for tok in doc:
    if tok.is_alpha and random.randrange(100) < chance:
        output.append(
            random.choice(list(similarsemantic(t, nlp, tok.text, nuttiness))) + tok.whitespace_)
    else:
        output.append(tok.text_with_ws)
print("".join(output))
```
