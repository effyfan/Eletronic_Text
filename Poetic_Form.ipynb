```python
import random
```

```python
initials = {}
```

```python
for item in lines:                      # loop through each item in words
    first_letter = item[0].lower()      # let first_letter = first letter of each item
    if first_letter not in initials:    # check if that first letter is alreay added as a key
        initials[first_letter] = []     # add first letters as keys with no value
    initials[first_letter].append(item) # add word to the key it belong to as a new value
```
