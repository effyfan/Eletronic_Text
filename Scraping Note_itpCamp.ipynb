## kitten tags

```python
kitten_tags = document.find_all("div", attrs={"class": "kitten"})
```

```python
kitten_tags[0]
```

```python
type(kitten_tags[0])
```

```python
type(kitten_tags)
```

```python
kitten_tags # a list of tag objects
```

```python
first_kitten = kitten_tags[0]
```

```python
first_kitten.find("h2")
```

```python
for item in kitten_tags:
    this_h2 = item.find("h2")
    print(this_h2.string)
```

```python
for item in kitten_tags:
    this_h2 = item.find("h2")
    print(this_h2.string)
    li_tags = item.find_all('li')
    print(li_tags)
```

```python
for item in kitten_tags:
    this_h2 = item.find("h2")
    print(this_h2.string)
    a_tags = item.find_all('a')
    print(a_tags)
```

```python
for item in kitten_tags:
    this_h2 = item.find("h2")
    print(this_h2.string)
    a_tags = item.find_all('a')
    a_tags_str = [t.string for t in a_tags]  # because there are 2 a_tags in each h2, we need for loop to make each one of them into string
    print(", ".join(a_tags_str))
```

```python
for item in kitten_tags:
    this_h2 = item.find("h2")
    kitten_name = this_h2.string
    a_tags = item.find_all('a')
    a_tags_str = [t.string for t in a_tags]
    kitten_shows = ", ".join(a_tags_str)
    print(kitten_name + ": " + kitten_shows)
```

## Data structure

list of dictionaries
    [
        {'name': Fluffy, 'shows': ['Deep Space Nine', 'Mr.Belvedere]}
        ...
    ]

```python
kitten_list = []

for item in kitten_tags:
    this_h2 = item.find("h2")
    kitten_name = this_h2.string
    a_tags = item.find_all('a')
    a_tags_str = [t.string for t in a_tags]
    kitten_shows = ", ".join(a_tags_str)
    this_kitten = {'name': kitten_name, 'shows': a_tags_str}
    kitten_list.append(this_kitten)
```

```python
kitten_list
```

```python
[item["name"] for item in kitten_list]
```

```python
reap2 = requests.get("https://tisch.nyu.edu/itp/news")
```

```python
html_str2 = reap2.text
```

```python
print(html_str2)
```

```python
 html_str2.find("media-content__wrap")   # not the beautiful soup find
```

```python
document = BeautifulSoup(html_str2, "html.parser")
```

```python
news_items = document.find_all("div", {"class": "media-content__wrap"})
```

```python
for item in news_items:
    blurb_tag = item.find("p", {"class": "media-item__content--maj"})
    print(blurb_tag)
```

```python
for item in news_items:
    first_a = item.find("a")
    title = first_a.string
    blurb_tag = item.find("p", {"class": "media-item__content--maj"})
    blurb = blurb_tag.string
    date_tag = item.find("p", {"class": "media-title--minor"})
    date = date_tag.string
    print(title, blurb, date)
```

## with image

```python
import time
```

```python
news_lis = document.find_all('li', {"class": "media-item--maj"})
```

```python
prefix = "https://tisch.nyu.edu"
for item in news_lis:
    img_tag = item.find("img")
    url = prefix + img_tag["src"]
    img_reap = requests.get(url)
    fname = url.split("/")[-1]
    open(fname, "wb").write(img_reap.content)
    time.sleep(0.5)
    print(fname)
```
