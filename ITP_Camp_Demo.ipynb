# Scrape Twitter

```python
reap2 = requests.get("https://twitter.com/itpcamp")
```

```python
html_str2 = reap2.text
```

```python
document2 = BeautifulSoup(html_str2, "html.parser")
```

```python
body_tag = document2.find("body", attrs = {"class", "three-col"})
```

```python
body_tag
```

```python
tweets = document2.find("li", attrs={"class", "js-stream-item"})
for tweet in tweets:
    print(tweet)
```

```python
tweet_text = tweets.find("p", attrs = {"class", "js-tweet-text"})
```

```python
tweet_text.text
```

### multipal tweets

```python
tweets = document2.find_all("li", attrs={"class", "js-stream-item"})
for tweet in tweets:
    print(tweet)
```

```python
for item in tweets:
    all_tweet_text = item.find("p", attrs = {"class", "js-tweet-text"})
    print(all_tweet_text.text + ";")
```

## Scrape NYU ITP NEWS

```python
reap3 = requests.get("https://tisch.nyu.edu/itp/news")
```

```python
html_str3 = reap3.text
```

```python
document3 = BeautifulSoup(html_str3, "html.parser")
```

```python
post_tags = document3.find_all("li", attrs={"class", "media-item--maj"})
```

```python
post_tags
```

```python
first_post = post_tags[0]
```

```python
first_post
```

```python
for item in post_tags:
    img_tag = item.find('img')
    url = "https://tisch.nyu.edu/" + img_tag["src"]
    img_reap = requests.get(url)
    filename = url.split("/")[-1]
    open(filename, "wb").write(img_reap.content)
    time.sleep(0.5)
    print(filename)
```
